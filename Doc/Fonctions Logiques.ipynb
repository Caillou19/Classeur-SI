Quelles étaient les raisons de cet hommage rendu par Google ?

l´hommage rendu par Google était dédié à George Boole, un mathématicien britannique, qui est l´inventeur de l'algèbre de boole. Cette date marquait le 200e anniversaire de la naissance de celui-ci. L'algèbre de boole est la base de la logique informatique moderne, utilisée pour la première fois pour les circuits de commutation téléphonique. 

Qu'est-ce qu'une variable booléenne ?

une variable qui peut prendre l'une des deux valeurs, True ou False

Qu'elles sont les principales fonctions (opérations) booléennes ?

AND (ET): La sortie est vraie uniquement si toutes les conditions sont vraies.
OR (OU): La sortie est vraie si au moins une condition est vraie.
NOT (NON) : La sortie est l'inverse de l'entrée (si l'entrée est vraie, la sortie devient fausse et vice versa).
XOR(OU exclusif) : La sortie est vraie uniquement si une seule des conditions est vraie, mais pas les deux.
